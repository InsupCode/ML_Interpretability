```python
base_stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    import re
    from nltk.tokenize import word_tokenize

    # Step 1: Clean text
    text = re.sub(r'[^a-zA-Z0-9\s/%±°]', '', text)
    text = text.lower()

    # Step 2: Tokenization
    doc = nlp(text)  # SpaCy tokenization
    tokens = [token.text for token in doc]  # Extract tokens from spaCy doc

    # Step 3: Identify medical terms using SciSpacy
    doc = nlp(" ".join(tokens))
    medical_terms = set(ent.text.lower() for ent in doc.ents)

    # Step 4: Create custom stop words dynamically
    custom_stop_words = base_stop_words - medical_terms

    # Step 5: Remove custom stop words from tokens
    filtered_tokens = [word for word in tokens if word not in custom_stop_words]

    # Step 6: Join tokens back into a string
    return " ".join(filtered_tokens)
```

### Tf-idf

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features=100)  # Using TF-IDF with a limit on the number of features
```

```python
def apply_tfidf(column_name, train_df, test_df, vectorizer):
    # Fit and transform the training data
    X_train_tfidf = vectorizer.fit_transform(train_df[column_name]).toarray()
    # Transform the test data using the already fitted vectorizer
    X_test_tfidf = vectorizer.transform(test_df[column_name]).toarray()

    # Create DataFrames with feature names based on the vectorizer's output
    tfidf_train = pd.DataFrame(X_train_tfidf, columns=[f'{column_name}_tfidf_{word}' for word in vectorizer.get_feature_names_out()])
    tfidf_test = pd.DataFrame(X_test_tfidf, columns=[f'{column_name}_tfidf_{word}' for word in vectorizer.get_feature_names_out()])

    # Concatenate the new features to the original train and test DataFrames
    train_df = pd.concat([train_df, tfidf_train], axis=1)
    test_df = pd.concat([test_df, tfidf_test], axis=1)

    return train_df, test_df
```

### Applying NER

```python
def extract_entities(text):
    doc = nlp(text)
    entities = [ent.text.lower() for ent in doc.ents]
    return entities
```

```python
title_train["entities"] = title_train["cleaned_title"].apply(extract_entities)
title_test["entities"] = title_test["cleaned_title"].apply(extract_entities)
```

### N-grams

```python
def apply_ngrams(column_name, train_df, test_df, ngram_range=(2, 3)):
    # Initialize the TF-IDF vectorizer with n-grams (2, 3)
    vectorizer = TfidfVectorizer(ngram_range=ngram_range, max_features=100)

    # Fit and transform the training data
    X_train_tfidf = vectorizer.fit_transform(train_df[column_name]).toarray()
    # Transform the test data using the already fitted vectorizer
    X_test_tfidf = vectorizer.transform(test_df[column_name]).toarray()

    # Create DataFrames with feature names based on the vectorizer's output
    ngram_train = pd.DataFrame(X_train_tfidf, columns=[f'{column_name}_ngram_{ngram}' for ngram in vectorizer.get_feature_names_out()])
    ngram_test = pd.DataFrame(X_test_tfidf, columns=[f'{column_name}_ngram_{ngram}' for ngram in vectorizer.get_feature_names_out()])

    # Concatenate the new features to the original train and test DataFrames
    train_df = pd.concat([train_df, ngram_train], axis=1)
    test_df = pd.concat([test_df, ngram_test], axis=1)

    return train_df, test_df
```

```python
title_train, title_test = apply_ngrams('cleaned_title', title_train, title_test, ngram_range=(2, 3))
```

```python
abs_train, abs_test = apply_ngrams('cleaned_abstract', abs_train, abs_test, ngram_range=(2, 3))
```
